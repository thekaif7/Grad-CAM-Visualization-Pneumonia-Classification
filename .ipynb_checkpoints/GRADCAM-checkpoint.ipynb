## GRAD-CAM  Pneumonia Classification 

Dataset from the below url

```python

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model, load_model
from IPython.display import display
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import matplotlib.cm as cm
```

```python
# re-size all the images to this
IMAGE_SIZE = [180, 180]
SIZE = 180
train_path = 'data/train'
valid_path = 'data/test'
```

```python
vgg = VGG16(input_shape=[SIZE,SIZE,3],include_top=False)
for layer in vgg.layers:
    layer.trainable = False
x = AveragePooling2D()(vgg.output)
x = Flatten()(x)
x = Dense(2, activation='softmax')(x)
model = Model(inputs=vgg.input, outputs=x)
model.summary()
```

```python
# tell the model what cost and optimization method to use
# model.compile(
#   loss='categorical_crossentropy',
#   optimizer='adam',
#   metrics=['accuracy']
# )
model.load_weights("model.h5")
```

```python
def get_img_array(img_path, size):
    # `img` is a PIL image of size 180180
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=size)
    # `array` is a float32 Numpy array of shape (180, 180, 3)
    array = tf.keras.preprocessing.image.img_to_array(img)
    # We add a dimension to transform our array into a "batch"
    # of size (1, 180, 180, 3)
    array = np.expand_dims(array, axis=0)
    return array
```

```python
labels = {1:"PNEUMONIA",0:"NORMAL"}
Normal = "data/test/NORMAL"
PNEUMONIA = "data/test/PNEUMONIA"

FilesP = [os.path.join(PNEUMONIA,i) for i in os.listdir(PNEUMONIA)]
FilesN = [os.path.join(Normal,i) for i in os.listdir(Normal)]

ORG = []
PRED = []
ImgFiles = []

for file in FilesP:
    img_array = get_img_array(file,(180,180))
    pred = model.predict(img_array)
    ImgFiles.append(file)
    ORG.append("Pneumonia")
    PRED.append(labels[np.argmax(pred[0])])

for file in FilesN:
    img_array = get_img_array(file,(180,180))
    pred = model.predict(img_array)
    ImgFiles.append(file)
    ORG.append("Normal")
    PRED.append(labels[np.argmax(pred[0])])

print(len(ImgFiles))
```

```python
def make_gradcam_heatmap(
    img_array, model, last_conv_layer_name, classifier_layer_names
):
    # First, we create a model that maps the input image to the activations
    # of the last conv layer
    last_conv_layer = model.get_layer(last_conv_layer_name)
    last_conv_layer_model = keras.Model(model.inputs, last_conv_layer.output)

    # Second, we create a model that maps the activations of the last conv
    # layer to the final class predictions
    classifier_input = keras.Input(shape=last_conv_layer.output.shape[1:])
    x = classifier_input
    for layer_name in classifier_layer_names:
        x = model.get_layer(layer_name)(x)
    classifier_model = keras.Model(classifier_input, x)

    # Then, we compute the gradient of the top predicted class for our input image
    # with respect to the activations of the last conv layer
    with tf.GradientTape() as tape:
        # Compute activations of the last conv layer and make the tape watch it
        last_conv_layer_output = last_conv_layer_model(img_array)
        tape.watch(last_conv_layer_output)
        # Compute class predictions
        preds = classifier_model(last_conv_layer_output)
        top_pred_index = tf.argmax(preds[0])
        top_class_channel = preds[:, top_pred_index]

    # This is the gradient of the top predicted class with regard to
    # the output feature map of the last conv layer
    grads = tape.gradient(top_class_channel, last_conv_layer_output)

    # This is a vector where each entry is the mean intensity of the gradient
    # over a specific feature map channel
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # We multiply each channel in the feature map array
    # by "how important this channel is" with regard to the top predicted class
    last_conv_layer_output = last_conv_layer_output.numpy()[0]
    pooled_grads = pooled_grads.numpy()
    for i in range(pooled_grads.shape[-1]):
        last_conv_layer_output[:, :, i] *= pooled_grads[i]

    # The channel-wise mean of the resulting feature map
    # is our heatmap of class activation
    heatmap = np.mean(last_conv_layer_output, axis=-1)

    # For visualization purpose, we will also normalize the heatmap between 0 & 1
    heatmap = np.maximum(heatmap, 0) / np.max(heatmap)
    
    
    
    return heatmap
```

```python
def get_GradCam_Img(heatmap,img_path):
    
    # We load the original image
    img = keras.preprocessing.image.load_img(img_path,target_size=(180,180))
    img = keras.preprocessing.image.img_to_array(img)

    # We rescale heatmap to a range 0-255
    heatmap = np.uint8(255 * heatmap)

    # We use jet colormap to colorize heatmap
    jet = cm.get_cmap("jet")

    # We use RGB values of the colormap
    jet_colors = jet(np.arange(256))[:, :3]
    jet_heatmap = jet_colors[heatmap]

    # We create an image with RGB colorized heatmap
    jet_heatmap = keras.preprocessing.image.array_to_img(jet_heatmap)
    jet_heatmap = jet_heatmap.resize((img.shape[1], img.shape[0]))
    jet_heatmap = keras.preprocessing.image.img_to_array(jet_heatmap)

    # Superimpose the heatmap on original image
    superimposed_img = jet_heatmap * 0.51 + img
    superimposed_img = keras.preprocessing.image.array_to_img(superimposed_img)
    
    return superimposed_img
```

```python
import random

last_conv = "block5_conv3"
classifier_layers = ["block5_pool","average_pooling2d","flatten_2","dense_1"]
fig, axs = plt.subplots(6,3, figsize = (16,32))
count = 0
for _ in range(6):
    i = random.randint(42, len(ImgFiles)-1)
    img_array = get_img_array(ImgFiles[i],(180,180))
    heatmap = make_gradcam_heatmap(img_array,model,last_conv,classifier_layers)
    
    img = tf.keras.preprocessing.image.load_img(ImgFiles[i], target_size=(180,180))
    
    axs[count][0].title.set_text(f"Original -{ORG[i]}")
    axs[count][0].imshow(img)
    
    axs[count][1].title.set_text("Heatmap") 
    axs[count][1].imshow(heatmap)
    
    gradcam = get_GradCam_Img(heatmap,ImgFiles[i])
    axs[count][2].title.set_text(f"Prediction - {PRED[i]}" ) 
    axs[count][2].imshow(gradcam)  
    
    count += 1

fig.tight_layout()
```
